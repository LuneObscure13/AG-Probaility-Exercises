## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
import random
import numpy as np
```

```python
# TODO

lanzamientos=1000
dado_1= np.random.randint(1,7, lanzamientos)
dado_2= np.random.randint(1,7, lanzamientos)

suma_total_dados= dado_1 + dado_2

ocurrencia_eventos= np.sum((suma_total_dados % 2 == 0)|(suma_total_dados > 7))

probabilidad_estimada=ocurrencia_eventos / lanzamientos

print(f"La probabilidad estimada de que el valor total sea par o mayor a 7 es de {probabilidad_estimada}")
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
ball_box = {}

# Crea la cajas con las bolas
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"


            
# TODO

first_event = 0      #Este contador es numeral para extraer 3 bolas blancas y 2 rojas
second_event = 0     #sacar todas del mismo color

for _ in range(1000): #numero de experimentos
    number_extractions = [ball_box[random.randint(0, 59)] for _ in range(5)] # numero de extracciones #agarra 5 bolas random de 60
    blancas = number_extractions.count("White") #Chequea number of extractions y dime cuantas veces sale x
    rojas = number_extractions.count("Red")
    verdes = number_extractions.count("Green")
    if blancas == 3 and rojas == 2:         #si se cumplen estas condiciones, TRUE
        first_event+=1                      #va contando cada vez hasta que se cumpla
    if blancas== 5 or rojas == 5 or verdes == 5:
        second_event+=1

first_event_prob= first_event/ 1000
second_event_prob= second_event / 1000
        
print(f"La probabilidad estimada de tomar 3 blancas y 2 rojas es {first_event_prob} y de tomar todas del mismo color es {second_event_prob}")
```
